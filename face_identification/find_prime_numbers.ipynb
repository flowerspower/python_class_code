```python
factors = []
def is_prime(number):
    if number <= 1:
        return False
    if number == 2:
        return True
    for i in range(number/2):
        if number % (i+2) == 0:
            return False
    return True

for i in range(1000):
    if is_prime(i):
        print i, "prime"
        
```
